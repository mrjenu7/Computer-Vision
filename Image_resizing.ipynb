```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('horse.jpg')
```

```python
image.shape
```

```python
target_size=(1500, 1200)
```

```python
resized_NNI = cv2.resize(image, target_size, interpolation=cv2.INTER_NEAREST)
resized_BILINEAR = cv2.resize(image, target_size, interpolation=cv2.INTER_LINEAR)
RESIZED_bicubic = cv2.resize(image, target_size, interpolation=cv2.INTER_CUBIC)
```

```python
resized_NNI.shape, resized_BILINEAR.shape, RESIZED_bicubic.shape
```

```python
#nn, bilinear, bicubic = cv2.resize(image, target_size)
```

```python
plt.figure(figsize=(16, 5))
plt.subplot(131)
plt.imshow(cv2.cvtColor(resized_NNI, cv2.COLOR_BGR2RGB))
plt.title('Nearest Neighbor Interpolation')
plt.axis('off')

plt.figure(figsize=(16, 5))
plt.subplot(132)
plt.imshow(cv2.cvtColor(resized_BILINEAR, cv2.COLOR_BGR2RGB))
plt.title('Bilinear Interpolation')
plt.axis('off')

plt.figure(figsize=(16, 5))
plt.subplot(133)
plt.imshow(cv2.cvtColor(RESIZED_bicubic, cv2.COLOR_BGR2RGB))
plt.title('Bicubic Interpolation')
plt.axis('off')

plt.tight_layout()
plt.show()
```
